```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Vinothkumar353-web/RAPIDO-Data-Analysis/refs/heads/main/rides_data.csv")
Rapido_df = pd.DataFrame(data)
Rapido_df.head()
```

```python
Rapido_df.rename(columns={"misc_charge":"Extra_Charges"},inplace=True)
Rapido_df.rename(columns={"services":"Vehicle_Type"},inplace=True)
Rapido_df
```

```python
Imputing_num_cols = ["ride_charge","Extra_Charges","total_fare"]


for col in Imputing_num_cols:
    Rapido_df[col] = Rapido_df.groupby("Vehicle_Type")[col].transform(lambda x: x.fillna(x.median()))
```

```python
Rapido_df['payment_method'] = Rapido_df['payment_method'].fillna(Rapido_df['payment_method'].mode()[0])
```

```python

Rapido_df['date'] = pd.to_datetime(Rapido_df['date']).dt.date


Rapido_df['time'] = Rapido_df['time'].astype(str).str.split('.').str[0]


Rapido_df['time'] = pd.to_datetime(Rapido_df['time'], format="%H:%M:%S").dt.time
```
